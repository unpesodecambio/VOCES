#  Código para la elaboración de la ficha

```python
import matplotlib.pyplot as plt
import pandas as pd

def graficar_datos_resaltados(gdf, columna, cve_met_interes, output_path=None):
    # Extraer datos de la columna
    datos = gdf[columna].sort_values().reset_index(drop=True)
    media = datos.mean()
    
    # Identificar la posición de la metrópoli de interés
    valor_interes = gdf.loc[gdf['CVE_MET'] == cve_met_interes, columna].values[0]
    posicion_interes = datos[datos == valor_interes].index[0]
    
    # Crear la gráfica
    fig, ax = plt.subplots(figsize=(5.31, 3.06))  # Ajustar las proporciones
    ax.plot(range(len(datos)), datos, color="lightgray", label="Datos")
    
    # Resaltar la posición de interés
    ax.scatter([posicion_interes], [valor_interes], color="blue", label=f'{cve_met_interes}', zorder=5)
    
    # Línea de la media
    ax.axhline(y=media, color="red", linestyle="--", linewidth=1.5, label="Media metropolitana")
    
    # Estilo
#    ax.set_title("Comparativa de Datos", fontsize=12, fontname="Century Gothic")
    ax.set_xlabel("Rango de Metrópolis", fontsize=8, fontname="Century Gothic")
    ax.set_ylabel(columna, fontsize=8, fontname="Century Gothic")
    ax.tick_params(axis="both", labelsize=7)
    ax.legend(fontsize=8, loc="upper left")
    ax.set_xlim([0, len(datos) - 1])
    
    # Guardar la gráfica si se especifica
    if output_path:
        plt.savefig(output_path, dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
# Sustituye 'loc' por tu GeoDataFrame y proporciona la columna de interés y CVE_MET.
columna = "DENS_EQ"
cve_met_interes = "01.1.01"

output_path = rf'Z:\VOCES\DATA\Tratados\NAC\FICHAS\Gráficas\{cve_met_interes}_{columna}.png'

graficar_datos_resaltados(
    gdf=loc, 
    columna=columna, 
    cve_met_interes=cve_met_interes, 
    output_path=output_path
)
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.drawing.image import Image
import matplotlib.pyplot as plt

# Cargar la base de datos
df = pd.read_excel('metropolis_data.xlsx')

# Cargar plantilla de Excel
template_path = 'ficha_template.xlsx'
wb = load_workbook(template_path)

for index, row in df.iterrows():
    # Crear una nueva hoja copiando la plantilla
    ws = wb.copy_worksheet(wb.active)
    ws.title = row['Metrópoli']

    # Insertar datos en las celdas predefinidas
    ws['B3'] = row['Nombre']
    ws['B4'] = row['Población']
    ws['B5'] = row['Indicador1']
    ws['B6'] = row['Indicador2']

    # Generar un gráfico de ejemplo
    plt.figure(figsize=(4, 3))
    plt.bar(['A', 'B', 'C'], [row['Indicador1'], row['Indicador2'], row['Indicador3']])
    plt.savefig('temp_chart.png')

    # Insertar gráfico en la hoja
    img = Image('temp_chart.png')
    ws.add_image(img, 'E5')

# Guardar el nuevo archivo con todas las fichas
wb.save('fichas_metropolis.xlsx')
```
